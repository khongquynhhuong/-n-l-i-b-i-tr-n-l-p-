```python
#Ví dụ: Lặp qua các khóa (key)
fruitscolor = {"Banana" : "Yellow",
"Mango" : "Green",
"Apple" : "Red",
"Grapefruit" : "Pink",
"Blackberry" : "Purple",
"Sapodilla" : "Brown"}
for key in fruitscolor:
    print (key)
```

```python
#Ví dụ: Lặp qua các giá trị (value)
fruitscolor = {"Banana" : "Yellow",
"Mango" : "Green",
"Apple" :
"Red",
"Grapefruit" : "Pink",
"Blackberry" : "Purple",
"Sapodilla" : "Brown"}
for color in fruitscolor.values():
    print(color)
```

```python
#Ví dụ: Lặp qua các cặp khóa và giá trị
fruitscolor = {"Banana" : "Yellow",
"Mango" : "Green",
"Apple" : "Red",
"Grapefruit" : "Pink",
"Blackberry" : "Purple",
"Sapodilla" : "Brown"}
for fruit, color in fruitscolor.items():
    print(fruit, ":", color)
```

```python
class MyNumbers:
    def __iter__(self):
        self.a = 1
        return self
    def __next__(self):
        x = self.a
        self.a += 1
        return x
myclass = MyNumbers()
myiter = iter(myclass)
print(next(myiter))
print(next(myiter))
print(next(myiter))
print(next(myiter))
```
